```python
import networkx as nx
import matplotlib.pyplot as plt
g = nx.Graph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,3)
g.add_edge(2,5)
g.add_edge(1,7)
g.add_edge(3,3)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(g,node_size=300,node_color="red",edge_color="blue",width=5)
plt.figsize = [150,150]
plt.show()
```
