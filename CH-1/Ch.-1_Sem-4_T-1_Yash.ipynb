### Outlier :

```python
import pandas as pd
import numpy as np

def find_outlier(ds,col):
    Q1 = ds[col].quantile(0.25)
    Q3 = ds[col].quantile(0.75)
    IQR = Q3 - Q1
    low_val = Q1 - (1.5*IQR)
    high_val = Q3 + (1.5*IQR)    
    ds = ds.loc[(ds[col]<low_val)|(ds[col]>high_val)]
    return ds

df = pd.read_csv('auto-mpg.csv')
find_outlier(df,"mpg")
```
